```python
# create figure and axis objects with subplots()
fig,ax = plt.subplots()
# make a plot
ax.plot(MWh_df.Year, MWh_df['Total End Use'], color="red", marker="o")
# set x-axis label
ax.set_xlabel("Year",fontsize=14)
# set y-axis label
ax.set_ylabel("Total Use - MWh", color="red",fontsize=14)

# twin object for two different y-axis on the sample plot
ax2=ax.twinx()
# make a plot with different y-axis using second axis object
ax2.plot(cents_KWh_df.Year, cents_KWh_df["All Sectors"],color="blue",marker="o")
ax2.set_ylabel("Cents per KWh",color="blue",fontsize=14)
plt.show()
# save the plot as a file
fig.savefig('two_different_y_axis_for_single_python_plot_with_twinx.jpg',
            format='jpeg',
            dpi=100,
            bbox_inches='tight')

```
