```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as stats
import requests
import json
from config import api_key
```

```python
#setting up API url

url = "http://api.eia.gov/series/?api_key="

elect_qua_df = pd.DataFrame(columns = ["State","2020","2019","2018","2017","2016","2015"])
states = ["AL","AK","AZ", "AR","CA","CO","CT", "DE","FL","GA", "HI", "ID", "IL","IN", "IA","KS", "KY","LA","ME","MD",
          "MA","MI","MN", "MS","MO","MT", "NE","NV","NH","NJ","NM","NY","NC","ND","OH","OK","OR","PA","RI","SC","SD",
          "TN","TX","UT","VT","VA","WA","WV","WI","WY"]


#looping through the call to get information calling API
for states in states:
    elec_df_q = requests.get(url+ api_key +"&series_id=ELEC.SALES."+states+"-ALL.Q").json()
    Q1 = elec_df_q['series'][0]['data'][1][1]
    Q2 = elec_df_q['series'][0]['data'][0][1]
    elec_df_y = requests.get(url+ api_key +"&series_id=ELEC.SALES."+states+"-ALL.A").json()
    yr_2020 = (Q1+Q2)/2
    new_row_df = {"State": states,"2020":yr_2020 ,
               "2019":elec_df_y['series'][0]['data'][0][1],
               "2018":elec_df_y['series'][0]['data'][1][1],
               "2017":elec_df_y['series'][0]['data'][2][1],
               "2016":elec_df_y['series'][0]['data'][3][1],
               "2015":elec_df_y['series'][0]['data'][4][1]
                 }
    elect_qua_df = elect_qua_df.append(new_row_df, ignore_index=True)

elect_qua_df.head(25)
```
